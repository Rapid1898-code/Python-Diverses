```python
# CSV-File einlesen
df =  pd.read_csv("ETH_1h.csv")
df.head(5)
```

```python
# Date steht als String zur Verfügung und nicht als DateTime
# Aufbau des Datums hat auch ein Sonderformat
# Daher muss das Format genau vorgegeben werden
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d %I-%p")
df.info()
```

```python
# Ausgabe des Wochentages für ein bestimmtes Datetime-Element
df.loc[0,"Date"].day_name()
```

```python
# Datetime-Umwandlung kann aber auc direkt beim CSV-Einlesen gemacht werden
# Umwandlung erfolgt hier mittels strptime()
d_parser =  lambda x: pd.datetime.strptime(x, "%Y-%m-%d %I-%p")
df =  pd.read_csv("ETH_1h.csv", parse_dates=["Date"], date_parser=d_parser)
```

```python
# Umwandlung der gesamten DateTimeSpalte auf den Wochentag pro Datum
df["Date"].dt.day_name()
```

```python
# Zusätzliche Spalte wird eingefügt mit dem Wochentag pro Zeile
df["DayOfWeek"] = df["Date"].dt.day_name()
df
```

```python
# Ältestes Datum wird ausgegeben
df["Date"].min()
```

```python
# Neuestes Datum wird ausgegeben
df["Date"].max()
```

```python
# Zeitdifferenz von 2 Datumswerten kann ermittelt werden
df["Date"].max() - df["Date"].min()
```

```python
# Es werden nur Sätze ausgegeben, die aus dem Jahr 2020 sind
filt=(df["Date"] >= "2020")
df.loc[filt]
```

```python
# Ausgabe von Sätzen aus 2019
filt=(df["Date"] >= "2019") & (df["Date"] < "2020")
df.loc[filt]
```

```python
# Ausgabe von Sätzen aus 2019 ab 30.6.2019
filt=(df["Date"] >= pd.to_datetime("2019-06-30")) & (df["Date"] < pd.to_datetime("2020-01-01"))
df.loc[filt]
```

```python
# Date als Index festlegen
df.set_index("Date", inplace=True)
```
